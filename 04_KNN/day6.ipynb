## Understanding Decision Boundaries in KNN

A decision boundary is a line or surface that divides different groups in a classification task. It shows which areas belong to which class based on what the model decides.

---

## Decision Boundaries in KNN

The shape of the boundary depends on:
- The value of K (how many neighbours are considered)
- How the data points are spread out in space

---

## Voronoi Diagrams

For a 1-nearest neighbour (1-NN) classifier, the decision boundary can be visualized using a Voronoi diagram.

A Voronoi diagram splits space into regions based on which training point is closest. Each region called a Voronoi cell contains all the points closest to one specific training point.

---

## How KNN Defines Decision Boundaries

### Impact of 'K' on Decision Boundaries

| k Value | Boundary Shape | Effect |
|---------|----------------|--------|
| Small k | Complex, irregular | Overfitting |
| Large k | Smooth, generalized | Underfitting |

### Impact of Distance Metric

| Metric | Boundary Shape |
|--------|----------------|
| Euclidean | Circular/elliptical boundaries |
| Manhattan | Axis-aligned boundaries |

---

## Factors Affecting KNN Decision Boundaries

| Factor | Effect |
|--------|--------|
| Feature Scaling | Features with larger ranges dominate distance calculations |
| Noise in Data | Outliers can shift or distort boundaries |
| Data Distribution | Influences how KNN separates classes |

---

## One-Line Summary

**KNN decision boundaries are determined by K value and distance metric, with small K creating complex boundaries and large K creating smoother ones.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

print("="*50)
print("KNN DAY 6 - DECISION BOUNDARIES (SCIKIT-LEARN)")
print("="*50)
```

```python
# Create dataset
X, y = make_classification(n_samples=200, n_features=2, n_informative=2,
                           n_redundant=0, n_clusters_per_class=1, random_state=42)

print(f"Dataset shape: {X.shape}")
print(f"Class 0: {np.sum(y == 0)}, Class 1: {np.sum(y == 1)}")
```

```python
# Plot decision boundaries for different k values
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
k_values = [1, 3, 5, 10]

for ax, k in zip(axs.flat, k_values):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X, y)
    
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    ax.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.Paired, marker='o', s=30)
    ax.set_title(f'KNN Decision Boundaries (k={k})')
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Compare different distance metrics
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
metrics = ['euclidean', 'manhattan']

for ax, metric in zip(axs, metrics):
    knn = KNeighborsClassifier(n_neighbors=3, metric=metric)
    knn.fit(X, y)
    
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    ax.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.Paired, marker='o', s=30)
    ax.set_title(f'Metric: {metric.capitalize()} (k=3)')
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Effect of feature scaling on decision boundaries
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Without scaling
X_unscaled = X.copy()
X_unscaled[:, 0] = X_unscaled[:, 0] * 10

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_unscaled, y)
Z = knn.predict(np.c_[xx.ravel()*10, yy.ravel()])
Z = Z.reshape(xx.shape)
axs[0].contourf(xx*10, yy, Z, alpha=0.3, cmap=plt.cm.Paired)
axs[0].scatter(X_unscaled[:, 0], X_unscaled[:, 1], c=y, edgecolor='k', cmap=plt.cm.Paired, s=30)
axs[0].set_title('Without Scaling')
axs[0].set_xlabel('Feature 1 (Scaled)')
axs[0].set_ylabel('Feature 2')
axs[0].grid(alpha=0.3)

# With scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_unscaled)

knn.fit(X_scaled, y)
x_min_s, x_max_s = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min_s, y_max_s = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx_s, yy_s = np.meshgrid(np.arange(x_min_s, x_max_s, 0.01), np.arange(y_min_s, y_max_s, 0.01))
Z = knn.predict(np.c_[xx_s.ravel(), yy_s.ravel()])
Z = Z.reshape(xx_s.shape)

axs[1].contourf(xx_s, yy_s, Z, alpha=0.3, cmap=plt.cm.Paired)
axs[1].scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, edgecolor='k', cmap=plt.cm.Paired, s=30)
axs[1].set_title('With Scaling')
axs[1].set_xlabel('Feature 1 (Scaled)')
axs[1].set_ylabel('Feature 2 (Scaled)')
axs[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
